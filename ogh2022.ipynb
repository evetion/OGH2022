```julia
url = "https://github.com/evetion/OGH2022/raw/main/ds801_small.csv"
```

```julia
using Downloads
using CSV, DataFrames
```

```julia
]add CSV DataFrames Plots
```

```julia
fn = Downloads.download(url)
```

```julia
@time df = DataFrame(CSV.File(fn, missingstring="NA"))
```

```julia
names(df)
```

```julia
filter!(row -> !ismissing(row.ba_ppm), df)
```

```julia
using Plots
scatter(df.longitude, df.latitude, marker_z=df.ba_ppm, clim=(0,1000), c=:deep)
```
